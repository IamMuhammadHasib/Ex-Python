```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use("ggplot")

x = [5, 9, 6]
y = [3, 6, 9]

x1 = [1, 0, 4]
y1 = [5, 5, 5]

plt.plot(x, y, 'g', label = 'line 1', linewidth = 5)
plt.plot(x1, y1, 'b', label = 'line 2', linewidth = 5)

plt.legend()

plt.title("Equation graph")
plt.xlabel("X")
plt.ylabel("Y")

plt.grid(True, color = 'k')

plt.show()
```

```python
plt.bar([1, 3, 5, 7, 9], [5, 2, 5, 7, 4], label = 'example 1', color = 'b')
plt.bar([2, 4, 6, 8, 10], [3, 9, 1, 5, 3], label = 'example 2', color = 'g')
plt.legend()
plt.xlabel("Bar")
plt.ylabel("Bar Height")
plt.show()
```
