```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use("ggplot")

x = [5, 9, 6]
y = [3, 6, 9]

x1 = [1, 0, 4]
y1 = [5, 5, 5]

plt.plot(x, y, 'g', label = 'line 1', linewidth = 5)
plt.plot(x1, y1, 'b', label = 'line 2', linewidth = 5)

plt.legend()

plt.title("Equation graph")
plt.xlabel("X")
plt.ylabel("Y")

plt.grid(True, color = 'k')

plt.show()
```
